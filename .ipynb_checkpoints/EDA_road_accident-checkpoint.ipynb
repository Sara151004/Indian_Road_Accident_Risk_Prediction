```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# load dataset
df = pd.read_csv("comprehensive_risk_indian_roads_dataset.csv")

# basic view
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'], dayfirst=True, errors='coerce')
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['festival'], inplace=True)
```

```python
df.drop(columns=['temperature'], inplace=True)
```

```python
df.drop(columns=['time'], inplace=True)
```

```python
df.drop(columns=['is_weekend_str'], inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
sns.countplot(
    x='accident_severity',
    hue='accident_severity',
    data=df,
    palette=['#FF9800','#4CAF50','#F44336'],
    legend=False
)

plt.title("Accident Severity Distribution")
plt.xlabel("Severity")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,8))

ax = sns.countplot(
    y='cause',
    data=df,
    order=df['cause'].value_counts().index,
    palette=['#FF3B30', '#FF9500', '#34C759', '#007AFF', '#AF52DE']
)

# add labels
for p in ax.patches:
    width = p.get_width()
    ax.text(width + 50, p.get_y() + p.get_height()/2,
            int(width), va='center')

plt.title("Major Causes of Accidents", fontsize=14, fontweight='bold')
plt.xlabel("Count")
plt.ylabel("Cause")
plt.show()
```

```python
plt.figure(figsize=(12,6))

ax = sns.countplot(x='hour', data=df, palette='viridis')

# add count labels
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2,
            height + 20,
            int(height),
            ha='center')

plt.title("Accidents by Hour", fontsize=14, fontweight='bold')
plt.xlabel("Hour of Day")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(
    x='is_weekend',
    y='risk_score',
    data=df,
    palette=['#2196F3', '#E91E63']
)

plt.xticks([0,1], ['Weekday','Weekend'])
plt.title("Weekend vs Risk Score")
plt.show()
```

```python
plt.figure(figsize=(10,6))

ax = sns.countplot(
    x='time_of_day',
    hue='accident_severity',
    data=df,
    palette=['#4CAF50', '#FF9800', '#F44336']
)

plt.title("Time of Day vs Accident Severity", fontsize=14, fontweight='bold')

# ✅ Add counts inside bars
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.text(
            p.get_x() + p.get_width()/2,
            height/2,
            int(height),
            ha="center",
            va="center",
            fontsize=9,
            fontweight='bold',
            color='black'
        )

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

corr = df[['hour','lanes','traffic_signal','vehicles_involved',
           'casualties','is_peak_hour','risk_score']].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    linewidths=0.5
)

plt.title("Correlation Heatmap (Key Factors)", fontsize=14, fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.violinplot(
    x='accident_severity',
    y='risk_score',
    data=df,
    palette=['#FF9800','#4CAF50','#F44336'],
    inner='quartile'
)
# Add count labels above each violin
counts = df['accident_severity'].value_counts()

for i, category in enumerate(counts.index):
    plt.text(i, df['risk_score'].max()+0.02,
             f'n={counts[category]}',
             ha='center', fontsize=10, fontweight='bold')
plt.title("Risk Score Distribution by Accident Severity", fontsize=14, fontweight='bold')
plt.xlabel("Severity")
plt.ylabel("Risk Score")
plt.show()
```

```python
sns.pairplot(
    df[['vehicles_involved', 'casualties', 'risk_score', 'hour', 'accident_severity']],
    hue='accident_severity',
    palette=['#4CAF50', '#FF9800', '#F44336'],
    diag_kind='kde'
)

plt.suptitle("Pairwise Relationship Between Key Variables", y=1.02, fontsize=14, fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='weather',
    hue='accident_severity',
    data=df,
    palette=['#4CAF50', '#FF9800', '#F44336']
)

plt.title("Weather vs Accident Severity", fontsize=14, fontweight='bold')
plt.xlabel("Weather Condition")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxenplot(
    x='traffic_density',
    y='risk_score',
    data=df,
    palette='coolwarm'
)

plt.title("Traffic Density vs Risk Score", fontsize=14, fontweight='bold')
plt.show()
```
